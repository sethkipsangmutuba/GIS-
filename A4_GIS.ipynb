# Conditional Statements

##Seth Kipsang

Conditional statements introduce **decision-making logic** into a program. Rather than executing instructions sequentially, Python can evaluate conditions and dynamically choose which path to follow. This capability is foundational for building intelligent systems, whether simple scripts or complex analytical pipelines.

---

## 1. The Logic of Conditions

At the core of conditional statements lies a simple principle:

> If a specified condition is true, execute a particular block of code; otherwise, execute an alternative.

This branching allows programs to respond differently depending on data values—essential in fields such as **epidemiological modeling, climate analysis, or policy simulations**.

---

## 2. Structure of Conditional Statements

Python uses three primary constructs:

- **if** → evaluates the first condition  
- **elif** → evaluates additional conditions if previous ones fail  
- **else** → executes when none of the conditions are true  

**Key structural features:**

- Each condition ends with a colon  
- The associated code block must be indented  
- Conditions are evaluated sequentially, stopping at the first true condition  

---

## 3. Boolean Logic and Evaluation

All conditions resolve to **Boolean values**:

- **True** → condition is satisfied  
- **False** → condition is not satisfied  

These Boolean outcomes determine program flow.  

In applied contexts such as a Nairobi-based climate-health dataset a condition might check whether temperature exceeds a threshold linked to heat stress. The Boolean result then determines subsequent actions, such as triggering alerts.

---

## 4. Comparison Operators

Conditional logic relies on comparison operators to evaluate relationships:

- Equality and inequality  
- Greater than / less than  
- Inclusive comparisons (≥, ≤)  

These operators form the backbone of **decision rules** in analytical workflows.

---

## 5. Chained Conditions with elif

Real-world decision-making rarely involves a single condition. The **elif** construct allows multiple, mutually exclusive conditions to be evaluated in sequence.

- Only the first true condition is executed; the rest are ignored  
- Ordering of conditions is critically important  

**Example logic:**

- Above threshold → one action  
- Exactly at threshold → another  
- Below threshold → a third  

Such structures are common in **classification problems and rule-based systems**.

---

## 6. Combining Conditions

Logical operators enable complex rules:

- **and** → all conditions must be true  
- **or** → at least one condition must be true  
- **not** → reverses the Boolean value  

**Example applications:**

- Environmental condition **and** health vulnerability  
- Rainfall **or** high wind speed triggering policy response  

These combinations are central to **risk modeling and decision-support systems**.

---

## 7. Sensitivity to Data Representation

Conditional statements are highly sensitive to **data representation**:

- Text comparisons are case-sensitive  
- Data types must be compatible  
- Small inconsistencies can produce incorrect evaluations  

This underscores the importance of **data preprocessing**, especially in applied contexts such as Kenyan administrative datasets or health registries.

---

## 8. Integrating Conditions with Loops

Conditional statements are often embedded within loops to evaluate multiple observations iteratively.

**Applications include:**

- Classification of datasets  
- Filtering of values  
- Automated decision-making across large datasets  

Example: looping through daily temperature records and classifying each as “hot” or “not hot” reflects basic **rule-based analytics**.

---

## 9. Key Takeaways

- Conditional statements enable decision-making in programs  
- They rely on Boolean logic and comparison operators  
- Multiple conditions can be structured using **if–elif–else** chains  
- Logical operators allow construction of complex rules  
- Integration with loops supports scalable data processing  

---

### Reflective Insight

Conditional statements mirror real-world decision processes—**public health triage, climate risk categorization, or behavioral choices**. Computationally, they form the bridge between **data and action**.

In advanced applications—such as AI-driven policy tools or early warning systems in East Africa—these simple constructs evolve into layered decision architectures. At their core, however, they remain grounded in **binary logic**: evaluating conditions and acting accordingly.

Understanding conditionals deeply is less about syntax and more about cultivating **structured reasoning under uncertainty**, a skill that transcends programming.

```python
import seaborn as sns
import pandas as pd

# Load the clean Iris dataset
iris = sns.load_dataset("iris")
print("Dataset head:\n", iris.head())
```

```python

# Simple if-else: classify sepal length
print("\n1. Simple if-else classification:")
for index, row in iris.iterrows():
    if row['sepal_length'] > 5.5:
        size_class = "Long Sepal"
    else:
        size_class = "Short Sepal"
    print(f"Sample {index}: {row['sepal_length']} cm → {size_class}")
```

```python
# Using if-elif-else: species and petal length classification
print("\n2. if-elif-else example:")
for index, row in iris.iterrows():
    if row['petal_length'] > 5:
        petal_size = "Large Petal"
    elif 3 < row['petal_length'] <= 5:
        petal_size = "Medium Petal"
    else:
        petal_size = "Small Petal"
    print(f"Sample {index}: {row['petal_length']} cm → {petal_size}")

```

```python
# Combining conditions with logical operators
print("\n3. Combining conditions (Sepal & Petal):")
for index, row in iris.iterrows():
    if row['sepal_length'] > 5.0 and row['petal_length'] > 4.0:
        category = "Large Sepal & Large Petal"
    elif row['sepal_length'] > 5.0 or row['petal_length'] > 4.0:
        category = "Either Large Sepal or Large Petal"
    else:
        category = "Small Sepal & Small Petal"
    print(f"Sample {index}: {category}")
```

```python
#Conditional statements integrated with new column
# Create a new column 'sepal_size_category' using conditions
iris_copy = iris.copy()
sepal_category = []

for i in range(len(iris_copy)):
    if iris_copy.loc[i, 'sepal_length'] > 5.5:
        sepal_category.append("Long")
    elif iris_copy.loc[i, 'sepal_length'] > 4.5:
        sepal_category.append("Medium")
    else:
        sepal_category.append("Short")

iris_copy['sepal_size_category'] = sepal_category

print("\n4. Sepal size category added (first 10 rows):")
print(iris_copy[['sepal_length', 'sepal_size_category']].head(10))

```

```python
#Nested conditions: sepal & petal categories
iris_copy['size_label'] = ""

for i in range(len(iris_copy)):
    if iris_copy.loc[i, 'sepal_length'] > 5.5:
        if iris_copy.loc[i, 'petal_length'] > 4.5:
            iris_copy.loc[i, 'size_label'] = "Large Sepal & Petal"
        else:
            iris_copy.loc[i, 'size_label'] = "Large Sepal & Small Petal"
    else:
        if iris_copy.loc[i, 'petal_length'] > 4.5:
            iris_copy.loc[i, 'size_label'] = "Small Sepal & Large Petal"
        else:
            iris_copy.loc[i, 'size_label'] = "Small Sepal & Petal"

print("\n5. Nested conditions applied (first 10 rows):")
print(iris_copy[['sepal_length', 'petal_length', 'size_label']].head(10))
```
