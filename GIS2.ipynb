## Introduction

Combining text and numerical values is a common task in programming. While simple concatenation using the `+` operator is possible, it becomes inefficient and error-prone for more complex expressions. Python provides structured approaches for string formatting that are more robust and readable.

## String Formatting Approaches

There are three main approaches:

1. **F-Strings (Recommended)**  
   - Modern and most intuitive method.  
   - Introduced in Python 3.6.  
   - Allows embedding variables directly inside text using `{}`.  
   - Automatically handles type conversion.  
   - Supports formatting (e.g., rounding numbers).  

   **Key features:**  
   - Prefix string with `f`  
   - Insert variables inside `{}`  
   - Control precision using format specifiers (e.g., `:.2f`)  

   *Concept:* Dynamically generate a sentence combining station name, ID, and temperature.

```python
# Sample data
station_name = "helsinki kaisaniemi"
station_id = 101
temperature = 12.3456
date = "2026-03-27"


# F-Strings (Recommended)

formatted_f = f"Station {station_name.title()} (ID: {station_id}) has temperature {temperature:.2f}°C"
print("1. F-string:", formatted_f)
```

2. **`.format()` Method (Legacy)**  
   - Uses placeholders inside `{}` within a string.  
   - Values are passed after the string.  
   - Flexible but less readable than f-strings.


```python

# .format() Method

formatted_format = "Station {} (ID: {}) has temperature {:.2f}°C".format(
    station_name.title(), station_id, temperature
)
print(".format():", formatted_format)
```

3. **`%` Operator (Deprecated Style)**  
   - Uses placeholders like `%s` and `%f`.  
   - Order-dependent and prone to errors.  
   - Not recommended for modern Python use.



```python

# % Operator (Deprecated)
formatted_percent = "Station %s (ID: %d) has temperature %.2f°C" % (
    station_name.title(), station_id, temperature
)
print("% formatting:", formatted_percent)

```

## String Manipulation Techniques

1. **Splitting Strings**  
   - `.split(character)` divides text into parts based on a delimiter.  
   - Returns a list of substrings.


```python
split_name = station_name.split(" ")
print("Split:", split_name)
```


2. **Slicing Strings**  
   - Extracts a portion of a string using index positions.  
   - Useful for cleaning or restructuring text.


```python
slice_example = station_name[:8]
print("Slice (first 8 chars):", slice_example)
```

3. **Replacing Text**  
   - `.replace(old, new)` substitutes parts of a string.  
   - Common for cleaning repeated or unwanted text.

```python
replaced_text = station_name.replace("helsinki", "Espoo")
print("Replace:", replaced_text)
```

4. **Changing Case**  
   - `.upper()` → uppercase  
   - `.lower()` → lowercase  
   - `.capitalize()` → first letter uppercase

```python
print("Upper:", station_name.upper())
print("Lower:", station_name.lower())
print("Capitalize:", station_name.capitalize())
```

5. **Concatenation**  
   - Strings can be combined using the `+` operator.  
   - Useful for constructing longer sentences from smaller parts.

```python
concat_text = "Station: " + station_name + ", Temp: " + str(temperature)
print("Concatenation:", concat_text)
```

6. **Formatting Text Width**  
   - `.zfill(n)` adds leading zeros to ensure fixed-width strings.  
   - Common in formatting dates or identifiers.

```python
formatted_id = str(station_id).zfill(5)
print("Zero-filled ID:", formatted_id)
```

7. **Multi-line Text**  
   - `\n` creates a new line within a string.  
   - Useful for structured outputs and reporting.

```python
multi_line = f"Station Summary:\nName: {station_name.title()}\nTemp: {temperature:.1f}°C"
print("Multi-line:\n", multi_line)
```

## Conceptual Example

**Goal:** Construct structured output such as:

- A summary sentence combining:  
  - Number of stations  
  - Station names (formatted consistently)  
- A second line containing a formatted date  

This involves:  

- Cleaning text (`split`, `replace`)  
- Formatting case (`uppercase`)  
- Formatting numbers (leading zeros)  
- Concatenating structured text

```python
stations = ["helsinki harmaja", "helsinki kumpula", "helsinki kaisaniemi"]

# Clean + format
stations_clean = [s.replace("helsinki", "").strip().upper() for s in stations]

summary = f"Total Stations: {len(stations_clean)} | Names: {', '.join(stations_clean)}"
date_formatted = f"Report Date: {date}"

final_output = summary + "\n" + date_formatted

print("Structured Output:\n", final_output)
```


## Key Takeaways

- Prefer f-strings for clarity and efficiency.  
- Strings are highly flexible and support many transformation methods.  
- Text processing is foundational in data science workflows, especially when preparing outputs, reports, and metadata.
