## Simple Python Math

Python can be used as a basic calculator. You can perform arithmetic operations such as addition, subtraction, multiplication, division, and exponentiation.

- **Addition:** 2 + 2  
- **Subtraction:** 4 - 2  
- **Multiplication:** 2 * 3  
- **Division:** 4 / 2  
- **Exponentiation:** 2 ** 3  

For more advanced math (e.g., square roots, trigonometry), Python requires modules such as `math`. Modules group related functions and constants (e.g., `math.pi`) that can be loaded using the import statement.

```python
# Basic Arithmetic Operations
print("Addition:", 2 + 2)
print("Subtraction:", 4 - 2)
print("Multiplication:", 2 * 3)
print("Division:", 4 / 2)
print("Exponentiation:", 2 ** 3)
```

## Functions

Functions are blocks of code that perform specific tasks.  

- Example: `print()` displays output, `math.sqrt()` calculates square roots.  
- Functions can be combined: you can use functions inside other functions to produce formatted outputs.

```python
import math

print("Value of pi:", math.pi)
print("sin(pi):", math.sin(math.pi))  # Expected ~0
```

## Variables

Variables store values for later use in a program.

- **Assignment:** variable_name = value  
- **Updating variables:** Values can be redefined anytime.  
- Variables can be used in expressions and combined with functions.  

**Example:**  

- Store a temperature in Celsius: temp_celsius = 10.0  
- Convert to Fahrenheit: tempFahrenheit = 9 / 5 * temp_celsius + 32  

Variable values persist across code cells after execution in environments like Jupyter Notebook.

```python

temp_celsius = 10.0
tempFahrenheit = 9 / 5 * temp_celsius + 32

print("Temperature in Celsius:", temp_celsius)
print("Temperature in Fahrenheit:", tempFahrenheit)
```


## Data Types

Python uses data types to define the characteristics of data:

| Data Type | Example   | Notes          |
|-----------|-----------|----------------|
| int       | 4         | Whole numbers  |
| float     | 3.1415    | Decimal numbers|
| str       | "Hot"     | Text/strings   |
| bool      | True      | Boolean values |

- Use `type(variable)` to check a variable’s data type.  
- Some operations require compatible data types. Incompatible types (e.g., float + str) will raise errors.

```python
# ==============================
# Data Types
# ==============================
a = 4
b = 3.1415
c = "Hot"
d = True

print("Type of a:", type(a))
print("Type of b:", type(b))
print("Type of c:", type(c))
print("Type of d:", type(d))

# ==============================
# Combining Text and Numbers
# ==============================
value = 25
print("The value is:", value)  # works
print("The value is " + str(value))  # explicit conversion

# ==============================
# String Operations
# ==============================
str1 = "Hello"
str2 = "World"

print("Concatenation:", str1 + " " + str2)
print("Repetition:", str1 * 3)
```

## Converting Between Data Types

Python provides functions to convert data:

- `float()` – converts a string or integer to a decimal number  
- `int()` – converts a string or float to an integer  
- `str()` – converts a number to a string  

> Be careful: converting floats to integers truncates the decimal part. Use `round()` to round if needed.

```python
num_str = "45.6"
num_float = float(num_str)

print("Converted to float:", num_float)
print("Calculation after conversion:", num_float + 10)

# Float to int (truncation)
print("Float to int:", int(num_float))
print("Rounded value:", round(num_float))
```

# Lists and Indices in Python

## Introduction to Lists

A list is a Python data type that stores multiple related values under a single variable.  

- Lists are defined using square brackets `[ ]`, with items separated by commas.  
- Lists can contain values of different types, but it is generally best practice to store values of the same type for data consistency.  

**Example:** A list of station names:  
station_names = ["Helsinki Harmaja", "Helsinki Kaisaniemi", "Helsinki Kaivopuisto", "Helsinki Kumpula"]



```python
# Creating a list
station_names = [
    "Helsinki Harmaja",
    "Helsinki Kaisaniemi",
    "Helsinki Kaivopuisto",
    "Helsinki Kumpula"
]

print("1. Station names:", station_names)

```

### Length of a List

- Use the `len()` function to determine how many items are in a list.



```python
# Length of list
print("2. Length:", len(station_names))
```

### Indexing

- Lists use zero-based indexing. The first item has index 0, the second 1, and so on.  
- Negative indices can be used to access items from the end of a list:  
  - `-1` gives the last item  
  - `-len(list)` gives the first item  
- Index values must be within the range of the list; otherwise, an IndexError occurs.  

**Analogy:** Think of a vending machine, where buttons (index values) select items (list values).



```python
print("3. First item (index 0):", station_names[0])
print("Second item (index 1):", station_names[1])
print("Last item (index -1):", station_names[-1])
print("First item using negative index:", station_names[-len(station_names)])

# IndexError example
try:
    print(station_names[10])
except IndexError as e:
    print("IndexError:", e)
```

### Modifying Lists

- Lists are mutable: their contents can be changed after creation.  
- Example modifications include updating a value at a given index or replacing a station type.



```python
station_names[0] = "Helsinki Updated"
print("Modified list:", station_names)
```

### Storing Multiple Data Types

- A list can hold different data types in the same collection: strings, integers, floats, etc.  
- **Example:** a single station’s information stored in a list: `[station_name, station_id, station_lat, station_lon, station_type]`


```python
station_info = [
    "Helsinki Kaisaniemi",  # name
    123,                    # id
    60.17,                  # latitude
    24.94,                  # longitude
    "Urban"                 # type
]

print("Mixed list:", station_info)
```


### Common List Operations

- **Adding items:** `.append(value)` adds an item to the end of the list.  
- **Removing items:**  
  - `del list[index]` removes an item at a specific index  
  - `.remove(value)` removes the first occurrence of a value  
  - `.pop(index)` removes and returns the item at a given index  
- **Counting and locating items:**  
  - `.count(value)` counts occurrences of a value  
  - `.index(value)` returns the index of the first occurrence of a value  
- **Reversing a list:** `.reverse()` reverses the order of items in-place  
- **Sorting a list:** `.sort()` sorts items alphabetically (capital letters come before lowercase letters)  

> Important: Some methods, like `.reverse()` and `.sort()`, modify the list in-place and return `None`. Do not assign their output back to the list variable.



```python
# Adding items
station_names.append("Helsinki New Station")
print("After append:", station_names)
```

```python
#Removing items

## Remove by index
del station_names[1]
print("After del:", station_names)
```

```python
# Remove by value
station_names.remove("Helsinki Kaivopuisto")
print("After remove:", station_names)
```

```python

# Pop (remove and return)
popped_item = station_names.pop(0)
print("Popped item:", popped_item)
print("After pop:", station_names)
```

```python
# Counting and locating
station_names.append("Helsinki Kumpula")
print("Count of 'Helsinki Kumpula':", station_names.count("Helsinki Kumpula"))
print("Index of 'Helsinki Kumpula':", station_names.index("Helsinki Kumpula"))

```

```python
# Reversing list
station_names.reverse()
print("Reversed list:", station_names)
```

```python
# Sorting list
station_names.sort()
print("Sorted list:", station_names)

# IMPORTANT: Demonstrating in-place behavior
result = station_names.sort()
print("Return value of sort():", result)  # None
print("List after sort():", station_names)
```
